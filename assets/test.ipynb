```python
import sqlite3 as sql

db=sql.connect('votingdb.db')
```

```python
cursor=db.cursor()
```

```python
sql = "SELECT name FROM sqlite_master WHERE type='table';"

# Execute the query
cursor.execute(sql)

# Fetch all rows from the result set
tables = cursor.fetchall()

# Print the table names
for table in tables:
    print(table[0])
```

```python
def convert(path):
    with open(path,'rb') as f:
        data=f.read()
        return data
```
